This program extracts the OCR result of each newspaper pages from the Los Angeles oil search results.

```python
import requests

# Change this to your own query
url = "https://www.loc.gov/newspapers/?end_date=1930-01-01&ops=~10&qs=los+angeles+oil&searchType=advanced&start_date=1892-04-20&location_country=united+states&fo=json"

# Call API
r = requests.get(url).json()


# Check how many results per page
print(r['results'])

# Example: Look at the 5th result's OCR
print(r['items'][4]['ocr_eng'][:500])  # preview first 500 characters
```
